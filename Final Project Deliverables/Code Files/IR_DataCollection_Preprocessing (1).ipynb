Following is the code for the collection of news data from the inshorts api, we have this code for current news data collection.

```python
import requests
#fetching the top news
news = []
url = 'https://inshortsv2.vercel.app/news/top'
response = requests.get(url)
#converting the resonse in the json format
x = response.json()
#printing the json format of the top news
x
```

```python
cnt = x['data']['count']
art = x['data']['articles']
#iterating over the total number of articles extracted above
for i in range (cnt):
    news = art[i]
    title = news['title']
    des = news['content']
    source = news['sourceName']
    source_url = news['sourceUrl']
    print('\n')
    print("TITLE: " + title)
    print("\nDESCRIPTION: "+des)
    print("\nSOURCE: "+source)
    print('\nSOURCE URL: '+source_url)
```

```python
#fetching the news topic wise, as an example: Entertainment. Similarly there are numerous other topics,
# namely: national,business,politics,sports,technology,startups,entertainment,hatke,education,world,automobile,science,travel,miscellaneous,fashion
news = []
url = 'https://inshorts.me/news/topics/entertainment'
response = requests.get(url)
#converting the resonse in the json format
z = response.json()
z
```

```python
cnt = z['data']['count']
art = z['data']['articles']
#iterating over the total number of articles extracted above
for i in range (cnt):
    news = art[i]
    title = news['title']
    des = news['content']
    source = news['sourceName']
    source_url = news['sourceUrl']
    print('\n')
    print("HEADLINE: " + title)
    print("\nDESCRIPTION: "+des)
    print("\nSOURCE: "+source)
    print('\nSOURCE URL: '+source_url)
```

```python
import pandas as pd  
  
# Calling DataFrame constructor  
dfE = pd.DataFrame(art)
#storing in the dataframe
dfE = dfE.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfE:
  dfE['categoryNames'] = 'ENTERTAINMENT'
dfE
```

```python
news = []
#national,business,politics,sports,technology,startups,entertainment,hatke,education,world,automobile,science,travel,miscellaneous,fashion
url = 'https://inshorts.me/news/topics/politics'
response = requests.get(url)
#converting the resonse in the json format
pol = response.json() #political data

url = 'https://inshorts.me/news/topics/national'
response = requests.get(url)
#converting the resonse in the json format
nat = response.json() #national data

url = 'https://inshorts.me/news/topics/business'
response = requests.get(url)
#converting the resonse in the json format
bus = response.json() #business data

url = 'https://inshorts.me/news/topics/sports'
response = requests.get(url)
#converting the resonse in the json format
sport = response.json() #sports data

url = 'https://inshorts.me/news/topics/technology'
response = requests.get(url)
#converting the resonse in the json format
tech = response.json() #tech data

url = 'https://inshorts.me/news/topics/education'
response = requests.get(url)
#converting the resonse in the json format
edu = response.json() #edu data

url = 'https://inshorts.me/news/topics/world'
response = requests.get(url)
#converting the resonse in the json format
wor = response.json() #world data

url = 'https://inshorts.me/news/topics/fashion'
response = requests.get(url)
#converting the resonse in the json format
fash = response.json() #fashion data
```

```python
#extracting the articles from the JSON data
artP = pol['data']['articles']
artF = fash['data']['articles']
artW = wor['data']['articles']
artE = edu['data']['articles']
artT = tech['data']['articles']
artS = sport['data']['articles']
artB = bus['data']['articles']
artN = nat['data']['articles']

#storing in the dataframe:
#storing the politics news data  
# Calling DataFrame constructor  
dfP = pd.DataFrame(artP)
#storing in the dataframe
dfP = dfP.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfP:
  dfP['categoryNames'] = 'POLITICS'
dfP

#storing the fashion news data
# Calling DataFrame constructor  
dfF = pd.DataFrame(artF)
#storing in the dataframe
dfF = dfF.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfF:
  dfF['categoryNames'] = 'FASHION'
dfF

#storing the world news data  
# Calling DataFrame constructor  
dfW = pd.DataFrame(artW)
#storing in the dataframe
dfW = dfW.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfW:
  dfW['categoryNames'] = 'WORLD'
dfW

#storing the Education news data
# Calling DataFrame constructor  
dfEd = pd.DataFrame(artE)
#storing in the dataframe
dfEd = dfEd.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfEd:
  dfEd['categoryNames'] = 'EDUCATION'
dfEd

#storing the technology news data  
# Calling DataFrame constructor  
dfT = pd.DataFrame(artT)
#storing in the dataframe
dfT = dfT.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfT:
  dfT['categoryNames'] = 'TECHNOLOGY'
dfT

#storing the sports news data
# Calling DataFrame constructor  
dfS = pd.DataFrame(artS)
#storing in the dataframe
dfS = dfS.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfS:
  dfS['categoryNames'] = 'SPORTS'
dfS

#storing the business news data  
# Calling DataFrame constructor  
dfB = pd.DataFrame(artB)
#storing in the dataframe
dfB = dfB.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfB:
  dfB['categoryNames'] = 'BUSINESS'
dfB

#storing the national news data
# Calling DataFrame constructor  
dfN = pd.DataFrame(artN)
#storing in the dataframe
dfN = dfN.drop(['hashId', 'oldHashId','authorName','sourceName','subtitle','imageUrl', 'shortenedUrl','relevancyTags','language','countryCode','score','important','createdAt'], axis=1)
for i in dfN:
  dfN['categoryNames'] = 'EDUCATION'
dfN
```

```python
# using ignore_index
res = pd.concat([dfE, dfP, dfP, dfN, dfS, dfB, dfF, dfEd, dfT], ignore_index=True)
res
```

```python
#renaming the dataframe columns as per Baseline.ipynb inout format
res.rename(columns = {'categoryNames':'category', 'title':'headlines', 'content':'short_description',
                              'sourceUrl':'link'}, inplace = True)
```

```python
#pickle the dataframe

  
import pickle
      
#creating a file
dbfile = open('NewsData', 'ab')
  
# source, destination
pickle.dump(res, dbfile)                     
dbfile.close()

# reading the file
dbfile = open('NewsData', 'rb')     
db = pickle.load(dbfile)
for keys in db:
    print(keys, '=>', db[keys])
dbfile.close()
  
```
